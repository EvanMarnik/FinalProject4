```python
import pandas as pd
import numpy as np
```

After importing Pandas, I imported the data set and checked what format it took.

```python
playground = pd.read_csv("playground-complete.csv")
playground.head(10)
```

Here I use value counts to count the amount of playgrounds for each neighborhood and take a sample of that query mask

```python
num_pgs = playground['neighborhood'].value_counts()
num_pgs.head(10)
```

```python
num_pgs.plot(kind="pie", label="")
```

Attempted to create a pie graph to represent the table but there are far too many neighborhoods. To simplify this, I made a visualization of a pie chart exclduing all neighboorhoods with only one playground

```python
multiple_pgs = num_pgs[num_pgs>1]
multiple_pgs.plot(kind="pie", label="")
```

To even further show the outlier here, we can use a bar graph

```python
multiple_pgs.plot.bar(x='n')
```
